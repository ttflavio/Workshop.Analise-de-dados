```python
caminho_arquivo = '/content/drive/MyDrive/dados_rh.csv'
df = pd.read_csv(caminho_arquivo)
```

```python
df
```

```python
# 1. Características do Banco de Dados
# a. Quantidade de linhas e colunas
print("Número de Linhas:", df.shape[0])
print("Número de Colunas:", df.shape[1])
```

```python
# b. Dados nulos
dados_nulos = df.isnull().sum()
print("Dados Nulos:")
print(dados_nulos)
```

```python
# c. Tipos de dados
tipos_dados = df.dtypes
print("\nTipos de Dados:")
print(tipos_dados)
```

```python
# 2. Agrupamento de Dados e Plotagem de Gráficos
# a. Gráfico de barras - Quantidade de funcionários por setor
df_department = df['Department'].value_counts()
df_department.plot(kind='bar', figsize=(8, 5), color='skyblue')
plt.title('Quantidade de Funcionários por Setor')
plt.xlabel('Setor')
plt.ylabel('Quantidade')
plt.xticks(rotation=45)
plt.show()
```

```python
# b. Gráfico de linha - Quantidade de funcionários por nível de escolaridade
df_education = df['Education'].value_counts().sort_index()
df_education.plot(kind='line', marker='o', figsize=(8, 5), color='green')
plt.title('Quantidade de Funcionários por Escolaridade')
plt.xlabel('Nível de Escolaridade')
plt.ylabel('Quantidade')
plt.grid()
plt.show()
```

```python
# c. Gráfico de pizza - Funcionários com ou sem atrito
df_attrition = df['Attrition'].value_counts()
df_attrition.plot(kind='pie', autopct='%1.1f%%', colors=['lightcoral', 'lightblue'], figsize=(6, 6))
plt.title('Percentual de Funcionários com ou sem Atrito')
plt.ylabel('')
plt.show()
```
